```python
import pandas as pd
```

```python
bc = pd.read_csv(r"C:\hp\Breast_Cancer.csv")
```

```python
bc.head(10)
```

```python
bc.shape
```

```python
bc.isnull().sum()
```

```python
bc.rename(columns ={'Cl.thickness': 'Cl_thickness' , 'Cell.size' : 'Cell_size' , 'Bare.nuclei': 'Bare_nuclei'}  , inplace = True)
bc.rename(columns ={'Cell.shape': 'Cell_shape' , 'Marg.adhesion' : 'Marg_adhesion' , 'Epith.c.size' : 'Epith_c_size'} , inplace = True)
bc.rename(columns ={'Bl.cromatin' : 'Bl_cromatin' ,  'Normal.nucleoli' : 'Normal_nucleoli'} , inplace = True)
```

```python
bc.head(10)
```

```python
bc.Bare_nuclei.fillna(bc.Bare_nuclei.mean()  , inplace = True)
```

```python
bc.isnull().sum()
```

```python
bc.select_dtypes(include = 'object').columns
```

```python
bc.Class.replace({'benign': 0 ,'malignant' :1 }  , inplace = True)
```

```python
bc.head(10)
```

```python
bc = bc.drop(['Unnamed: 0' , 'Id']  , axis = 1)
```

```python
bc.head(10)
```

```python
from sklearn.model_selection  import train_test_split
bc_train ,bc_test = train_test_split(bc , test_size = .2)
```

```python
bc_train_x = bc_train.iloc[: , 0:-1]
bc_train_y = bc_train.iloc[: , -1]
```

```python
bc_test_x = bc_test.iloc[: , 0:-1]
bc_test_y = bc_test.iloc[: , -1]
```

```python
from sklearn.linear_model  import   LogisticRegression 

logreg = LogisticRegression()

logreg.fit ( bc_train_x  ,  bc_train_y)
```

```python
pred = logreg.predict(bc_test_x)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tab1 = confusion_matrix(bc_test_y, pred )
```

```python
tab1
```

```python
tab1.diagonal().sum() *  100 / tab1.sum()
```

```python
from sklearn.metrics  import recall_score
recall_score(bc_test_y  ,  pred) * 100
```

```python
from sklearn.metrics  import f1_score
f1_score(bc_test_y  ,  pred) * 100
```

```python
from sklearn.metrics  import precision_score
precision_score(bc_test_y  ,  pred) * 100
```
